# Make our icon

```python
import meshio
from naivefea.constitutive import LinearMaterial
from naivefea.analysis import LinearFea
```

```python
mesh=meshio.read('makeicon.inp')
fea=LinearFea(mesh)
fea.set_figsize('large')
fea.plot_mesh(node=True,element=True)
```

```python
text_region=mesh.cell_sets['textregion'][0]
```

```python
rubber=LinearMaterial(10.0,0.3)
rubber.set_name('rubber')
enhance=LinearMaterial(50.0,0.3)
enhance.set_name('enhance')
fea.uniform_material(rubber)
fea.uniform_material(enhance,text_region)
fea.uniform_material(rubber,[108,109])
```

```python
fea.plot_material()
```
